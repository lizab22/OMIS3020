```python
import os
import subprocess
from google.colab import drive
import getpass

# Mount Google Drive
drive.mount("/content/drive")

# Repository settings (EDIT THESE)
repository = ""        # enter your repository name
username = ""       # enter your GitHub username
email = "@gmail.com"  # enter your GitHub email (or noreply email)
name = ""     # enter your full name (for commit history)

# Enter your GitHub token (hidden input)
token = getpass.getpass("Paste your GitHub token (hidden): ").strip()

# Where you want the repo to live in Drive
folder_path = f"/content/drive/MyDrive/{repository}"

# Go to MyDrive
%cd /content/drive/MyDrive

# Clone only if needed
if not os.path.exists(folder_path):
    clone_url = f"https://x-access-token:{token}@github.com/{username}/{repository}.git"
    !git clone "{clone_url}"
else:
    print(f"Repo already exists: {folder_path}")

# Move into the repo
%cd "{folder_path}"

# Set Git identity (global for this Colab runtime)
subprocess.run(["git", "config", "--global", "user.email", email], check=True)
subprocess.run(["git", "config", "--global", "user.name", name], check=True)
```
